```python
from qiskit import Aer, QuantumCircuit, transpile
from qiskit.visualization import plot_histogram, plot_state_qsphere
```

```python
def get_simulator_results(circuit):
    """
     Returns the simulator results for a circuit. This is a convenience function to use in conjunction with : func : ` get_stats `.
     
     @param circuit - The circuit to simulate. Must be a list of ` ` ( name value ) ` ` tuples.
     
     @return A list of ` ` ( name value count ) ` ` tuples where ` ` name ` ` is the name of the metric and ` ` count ` ` is the number of times it took to run
    """
    simulator = Aer.get_backend("aer_simulator")
    compiled_circuit = transpile(circuit, simulator)
    result = simulator.run(compiled_circuit).result()
    return result.get_counts()


def get_circuit_statevectors(circuit):
    """
     Given a circuit run it through Aer and return statevectors. This is a convenience function for simulating states and transitions in the circuit
     
     @param circuit - circuit to simulate. It must be a Circuit
     
     @return list of statevectors that were
    """
    simulator = Aer.get_backend("statevector_simulator")
    compiled_circuit = transpile(circuit, simulator)
    result = simulator.run(compiled_circuit).result()
    return result.get_statevector()
```

## 1st Circuit


```python
circuit = QuantumCircuit(2)

circuit.x(1)
circuit.cx(0, 1)
circuit.swap(0, 1)
circuit.measure_all()
circuit.draw("mpl")
```

Bloch spheres

```python
plot_state_qsphere(get_circuit_statevectors(circuit))
```

```python
plot_histogram(get_simulator_results(circuit))
```
